```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv(r"D:\Downloads\house price prediction in india.csv")
```

```python
# Display the first few rows
print(df.head())
```

```python
# UNIVARIATE ANALYSIS
# First analyse numerical values

# Summary statistics of numerical variables
print(df.describe())
```

```python
# Visualizing the distribution of Price

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price (in INR lakh)')
plt.ylabel('Frequency')
plt.show()

#The price distribution is right-skewed, indicating that most properties are priced lower with a few high-value listings.
```

```python
# Visualizing the distribution of Price per Sqft

plt.figure(figsize=(10, 6))
sns.histplot(df['price_per_sqft'], bins=30, kde=True)
plt.title('Price per Sqft Distribution')
plt.xlabel('Price per Sqft (in INR)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the distribution of Area

plt.figure(figsize=(10, 6))
sns.histplot(df['area'], bins=30, kde=True)
plt.title('Area Distribution')
plt.xlabel('Area (in sq. ft.)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Now analyse categorical variables

# Count of properties by property type
plt.figure(figsize=(10, 6))
sns.countplot(x='facing', data=df)
plt.title('Count of fancing direction')
plt.xlabel('fancing direction')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

#The direction a property faces also influence its price, depending on the market preferences east facing property is in more demand.
```

```python
# Count of properties by society
plt.figure(figsize=(10, 6))
sns.countplot(x='society', data=df, order=df['society'].value_counts().index[:10])  # Top 10 societies
plt.title('Count of Properties by Society')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count of properties by No. of bedRoom 
plt.figure(figsize=(10, 6))
sns.countplot(x='bedRoom', data=df)
plt.title('Count of No. of bedRoom ')
plt.xlabel('No. of bedRoom ')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count of properties by No. of bathroom

plt.figure(figsize=(10, 6))
sns.countplot(x='bathroom', data=df)
plt.title('Count of property having given No. of bathroom ')
plt.xlabel('No. of bathroom ')
plt.ylabel('Count of properties')
plt.show()
```

```python
# Count of properties by floorNum

plt.figure(figsize=(10, 6))
sns.countplot(x='floorNum', data=df)
plt.title('Count of properties on given floor number')
plt.xlabel('floorNum ')
plt.ylabel('Count of properties')
plt.show()
```

```python
#MULTIVARIATE ANALYSIS

# Numerical vs. Categorical

# Boxplot of Price by Property Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='property_type', y='price', data=df)
plt.title('Price vs Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price (in INR)')
plt.xticks(rotation=45)
plt.show()

#Different property types show significant variations in prices, indicating that property type is a critical factor.
```

```python
# Boxplot of Price by Society
plt.figure(figsize=(10, 6))
sns.boxplot(y='facing', x='price', data=df)
plt.title('Price vs Facing')
plt.xlabel('Price (in INR)')
plt.ylabel('Facing')
plt.show()
```

```python
# Correlation Analysis
# Calculate the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Showing scatter plot

# Scatter plot of Price vs Area
plt.figure(figsize=(10, 6))
sns.scatterplot(x='area', y='price', data=df)
plt.title('Price vs Area')
plt.xlabel('Area (in sq. ft.)')
plt.ylabel('Price (in INR)')
plt.show()
```

```python
# Scatter plot of Price vs Price per Sqft
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price_per_sqft', y='price', data=df)
plt.title('Price vs Price per Sqft')
plt.xlabel('Price per Sqft (in INR)')
plt.ylabel('Price (in INR lakh)')
plt.show()
```
